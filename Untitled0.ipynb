```python
from sklearn.model_selection import train_test_split
saleData['price'].replace(0, np.nan, inplace=True)
y = saleData['price'].fillna(value=saleData['price'].mean())
# all good to go
x = saleData.drop('price', axis=1)
x = x.drop('country', axis=1)
x = x.drop('street', axis=1)
x = x.drop('statezip', axis=1)
x = x.drop('yr_built', axis=1)
x = x.drop('yr_renovated', axis=1)
x = x.drop('city',axis=1)
x = x.drop('date', axis=1)
x = x.drop('view', axis=1)
x = x.drop('waterfront', axis=1)
x = x.drop('sqft_basement', axis=1)



x = np.array(x.values)
y = np.array(y.values)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
xtrain
```

```python
lr = LinearRegression()
lr.fit(xtrain, ytrain)
predlr = lr.predict(xtest)
scorelr=r2_score(ytest,predlr)
scorelr
```

```python
from sklearn import linear_model
ridge = linear_model.Ridge()
ridge.fit(xtrain,ytrain)
pred = ridge.predict(xtest)
r2_score(ytest,pred)
```

```python
rf = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs=-1, max_depth=35,max_features = 'sqrt',min_samples_split = 10)

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

pipe = make_pipeline(scaler,rf)

pipe.fit(xtrain, ytrain)

print("Random Forest Score",pipe.score(xtest,ytest))
```
